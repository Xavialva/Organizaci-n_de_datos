la función *'melt'* sirve para ordenar las columnas en filas, en algunos casos sirve para ordenar el dataframe. Esta función con id_vars ancla las columnas que no van hacer afectadas por la función.

```python
df1=df.melt(id_vars=['Country Name','Country Code','Indicator Name','Indicator Code'])
```

```python
df2=df1.drop(columns=['Indicator Name','Indicator Code'])
```

```python
df2
```

```python
df3=df2.rename(columns={'variable': 'Years','value':'Pib'})
```
